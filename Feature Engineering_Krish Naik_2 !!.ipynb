### Ordinal Number Encoding !!


```python
import datetime
today_date = datetime.datetime.today()
today_date
```

```python
today_date - datetime.timedelta(3)
```

```python
#list Comprehensions:
days = [today_date - datetime.timedelta(x) for x in range(0,15)]
```

```python
import pandas as pd
data = pd.DataFrame(days, columns = ["Day"])
data.head()
```

```python
data["weekday"] = data["Day"].dt.day_name()
data.head()
```

```python
dictionary = {"Monday":1,"Tuesday":2,"Wednesday":3,"Thursday":4,"Friday":5}
dictionary
```

```python
data["Weekday_oridnal"] = data["weekday"].map(dictionary)
data.head()
```

### Count of Frequency Encoding !!

```python
train_set = pd.read_csv("adult.csv")
train_set.head()
```

```python
for feature in train_set.columns[:]:
    print(feature,":",len(train_set[feature].unique()), "labels")  ##use the number of categories and replace it with the same values.
```

```python
country = train_set["native.country"].value_counts().to_dict()
```

```python
train_set["country"] = train_set["native.country"].map(country)
train_set.head()
```

```python
train_set.drop("native.country", axis = 1)
```

```python
# Adv: easy to use, not increasing any feature space, 
# Dis: it will not handle if the categories are same. it will provide the same corresponding count.
```
